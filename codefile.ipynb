```python
X=[]
y=[]
```

```python
for sq in seq:
  for i in range(1,len(sq)):
    input_seq=sq[:i]
    output_seq=sq[i]
    X.append(input_seq)
    y.append(output_seq)
```

```python
maxLen=max(len(x) for x in X)
print(maxLen)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
X_padded=pad_sequences(X,maxlen=maxLen,padding='pre')
```

```python
y=np.array(y)
```

```python
X_padded.shape
```

```python
from tensorflow.keras.utils import to_categorical
y_one_hot=to_categorical(y,num_classes=vocab_size)
```

```python
y_one_hot.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding , SimpleRNN, LSTM
```

```python
embed_dim=50
rnn_units=128
```

```python
rnn_model=Sequential()
```

```python
rnn_model.add(
    Embedding(input_dim=vocab_size,output_dim=embed_dim,input_length=maxLen)
)
```

```python
rnn_model.add(SimpleRNN(units=rnn_units))
rnn_model.add(Dense(units=vocab_size,activation='softmax'))
```

```python
rnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
rnn_model.summary()
```

```python
lstm_model=Sequential()
lstm_model.add(
    Embedding(input_dim=vocab_size,output_dim=embed_dim,input_length=maxLen)
)
lstm_model.add(LSTM(units=rnn_units))
lstm_model.add(Dense(units=vocab_size,activation='softmax'))
```

```python
lstm_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
from tensorflow.keras.models import load_model
lstm_model=load_model('/content/lstm_model.h5')
```

```python
index_to_word=tok.index_word
index_to_word
```

```python
def predictor(model,tok,text,maxlen):
  text=text.lower()

  seq=tok.texts_to_sequences([text])[0]
  seq=pad_sequences([seq],maxlen=maxlen,padding='pre')

  pred=model.predict(seq,verbose=0)
  pred_word=index_to_word[np.argmax(pred)]

  return pred_word
```

```python
seed_text="life is"
next_word=predictor(lstm_model,tok,seed_text,maxLen)
print(next_word)
```

```python
def generate_text(model,tok,seed_text,maxLen,n_words):
  for _ in range(n_words):
    next_word=predictor(model,tok,seed_text,maxLen)
    if next_word == "":
      break
    seed_text += " " + next_word

  return seed_text
```

```python
seed="the meaning of life"
generated_text=generate_text(lstm_model,tok,seed,maxLen,5)
```
